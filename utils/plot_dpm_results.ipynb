```python
import numpy as np
import os

def plot_a_metric(metric_name, metrics_data_dict, output_dir, dataset_name):
    # dataset: ego_small
    # metric: degree
    # method: sampler_euler_none
    if "unique" in metric_name:
        y_axis_sampler_euler_none = metrics_data_dict['sampler_euler_none']['unique@1000'] + metrics_data_dict['sampler_euler_none']['unique@999']
        y_axis_sampler_euler_langevin = metrics_data_dict['sampler_euler_langevin']['unique@512']
        y_axis_sampler_reverse_none = metrics_data_dict['sampler_reverse_none']['unique@512'] + metrics_data_dict['sampler_reverse_none']['unique@511']
        y_axis_sampler_reverse_langevin = metrics_data_dict['sampler_reverse_langevin']['unique@512']
        y_axis_sampler_s4_none = metrics_data_dict['sampler_s4_none']['unique@512'] 
        y_axis_sampler_dpm_none = metrics_data_dict['sampler_dpm_none']['unique@512']
        y_axis_sampler_dpm_langevin = metrics_data_dict['sampler_dpm_langevin']['unique@1000']
    else:
        y_axis_sampler_euler_none = metrics_data_dict['sampler_euler_none'][metric_name]
        y_axis_sampler_euler_langevin = metrics_data_dict['sampler_euler_langevin'][metric_name]
        y_axis_sampler_reverse_none = metrics_data_dict['sampler_reverse_none'][metric_name]
        y_axis_sampler_reverse_langevin = metrics_data_dict['sampler_reverse_langevin'][metric_name]
        y_axis_sampler_s4_none = metrics_data_dict['sampler_s4_none'][metric_name]
        y_axis_sampler_dpm_none = metrics_data_dict['sampler_dpm_none'][metric_name]
        y_axis_sampler_dpm_langevin = metrics_data_dict['sampler_dpm_langevin'][metric_name]

    x_axis = [10, 15, 20, 50, 200, 500, 1000]


    # plot the graph with label: sampler_euler_none and title: Ego Small
    import matplotlib.pyplot as plt

    # set the resolution of the figure
    plt.figure(dpi=300)
    # plt.title(dataset_name)
    plt.xlabel("NFE")
    # captalize the first letter of the metric name
    if "unique" in metric_name:
        metric_name = "unique@1000"
        
    metric_name = metric_name[0].upper() + metric_name[1:]
    plt.ylabel(' '.join(metric_name.split('_')))

    plt.plot(x_axis, y_axis_sampler_euler_none, '-x', label = "Euler")
    plt.plot(x_axis, y_axis_sampler_euler_langevin, '-x', label = "Euler_langevin")
    plt.plot(x_axis, y_axis_sampler_reverse_none, '-x', label = "Reverse")
    # plt.plot(x_axis, y_axis_sampler_reverse_langevin, '-x', label = "reverse_langevin")
    plt.plot(x_axis, y_axis_sampler_s4_none, '-x', label = "S4")
    plt.plot(x_axis, y_axis_sampler_dpm_none, '-x', label = "DPM")
    leg = plt.legend(prop={"size":15})
    # plt.plot(x_axis, y_axis_sampler_dpm_langevin, '-x', label = "dpm_langevin")

    # show label for the legend
    # plt.legend()
    plt.tight_layout()

    # save the figure
    dataset_name = '_'.join(dataset_name.lower().split())
    ouput_file = os.path.join(output_dir, dataset_name + "_" + metric_name + ".png")
    plt.savefig(ouput_file)
```
